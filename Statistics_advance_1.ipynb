1. A random variable is a numerical representation of the possible outcomes of a random process. It assigns numerical values to events in a probability experiment.

For example:

In a dice roll, the number appearing on the face (1 to 6) is a random variable.

In a coin flip, "Heads" could be assigned the value 1, and "Tails" the value 0.

Random variables are crucial because they allow probability theory to quantify uncertainty mathematically.

2. Types of Random Variables
Random variables fall into two main categories:

Discrete Random Variables

Can take only specific, countable values.

Example: The number of emails received in a day.

Common Distributions: Binomial, Poisson.

Continuous Random Variables

Can take any value within a certain range.

Example: The exact height of a person, temperature readings.

Common Distributions: Normal, Uniform.

3. Difference Between Discrete and Continuous Distributions
Discrete Distribution: Probability is assigned to distinct, separate values (e.g., rolling a die where only 1, 2, 3, 4, 5, or 6 can occur).

Continuous Distribution: Probability is spread over a range of values (e.g., a person's height can be 170.1 cm, 170.11 cm, 170.111 cm, etc.).

Mathematically:

Discrete uses probability mass functions (PMF).

Continuous uses probability density functions (PDF).

```python
# prompt: Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt

# Define the range of possible outcomes for the discrete uniform distribution
outcomes = np.arange(1, 7) # For example, rolling a fair six-sided die

# Calculate the probability for each outcome
# For a discrete uniform distribution, the probability of each outcome is 1 / (number of outcomes)
n_outcomes = len(outcomes)
probabilities = np.full(n_outcomes, 1 / n_outcomes)

# Print the outcomes and their probabilities
print("Outcomes:", outcomes)
print("Probabilities:", probabilities)

# Plot the Probability Mass Function (PMF)
plt.figure(figsize=(8, 6))
plt.bar(outcomes, probabilities, color='skyblue')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function of a Discrete Uniform Distribution')
plt.xticks(outcomes)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# prompt:  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
  """
    Calculates the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Args:
      x: The outcome (0 for failure, 1 for success).
      p: The probability of success.

    Returns:
      The probability of the given outcome.
    """
  if x == 1:
    return p
  elif x == 0:
    return 1 - p
  else:
    return 0 # For outcomes other than 0 or 1

# Example usage:
p_success = 0.7  # Probability of success
outcome_success = 1
outcome_failure = 0

prob_success = bernoulli_pdf(outcome_success, p_success)
prob_failure = bernoulli_pdf(outcome_failure, p_success)

print(f"Probability of success ({outcome_success}) with p={p_success}: {prob_success}")
print(f"Probability of failure ({outcome_failure}) with p={p_success}: {prob_failure}")
```

```python
# prompt: Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
# uniform distribution

# Calculate the Cumulative Distribution Function (CDF)
# The CDF for a discrete uniform distribution at a point x is the sum of probabilities for all outcomes less than or equal to x.
# For a discrete uniform distribution from 'a' to 'b' (inclusive), the CDF is:
# F(x) = 0 if x < a
# F(x) = (floor(x) - a + 1) / (b - a + 1) if a <= x <= b
# F(x) = 1 if x > b

# Let's use the same outcomes as the PMF example (1 to 6)
outcomes_cdf = outcomes
n_outcomes_cdf = len(outcomes_cdf)

# Calculate the CDF values for each outcome
cdf_values = [(i + 1) / n_outcomes_cdf for i in range(n_outcomes_cdf)]

# Plot the Cumulative Distribution Function (CDF)
plt.figure(figsize=(8, 6))
# We plot the CDF as a step function for a discrete distribution
plt.step(outcomes_cdf, cdf_values, where='post', color='forestgreen')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function of a Discrete Uniform Distribution')
plt.xticks(outcomes_cdf)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
# prompt: Write a Python function to calculate Z-scores from a dataset and plot them

def calculate_and_plot_zscores(data):
  """
  Calculates Z-scores for a dataset and plots a histogram of the Z-scores.

  Args:
    data: A list or NumPy array of numerical data.
  """
  if not isinstance(data, np.ndarray):
    data = np.array(data)

  if len(data) < 2:
    print("Dataset must contain at least two elements to calculate Z-scores.")
    return

  mean = np.mean(data)
  std_dev = np.std(data)

  if std_dev == 0:
      print("Standard deviation is zero. Z-scores cannot be calculated.")
      return

  z_scores = (data - mean) / std_dev

  print("Original Data:", data)
  print("Mean:", mean)
  print("Standard Deviation:", std_dev)
  print("Calculated Z-scores:", z_scores)

  # Plot the histogram of Z-scores
  plt.figure(figsize=(10, 6))
  plt.hist(z_scores, bins=20, density=True, edgecolor='black', alpha=0.7)
  plt.xlabel('Z-score')
  plt.ylabel('Density')
  plt.title('Histogram of Z-scores')
  plt.grid(axis='y', linestyle='--', alpha=0.7)
  plt.show()

# Example usage:
sample_data = [10, 12, 12, 13, 12, 11, 14, 13, 15, 10, 10, 12, 14, 14, 11, 16, 13, 10, 11, 12]
calculate_and_plot_zscores(sample_data)

sample_data_2 = np.random.normal(50, 10, 100) # Generate data from a normal distribution
calculate_and_plot_zscores(sample_data_2)
```

```python
# prompt: Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

# Choose a non-normal distribution. A Poisson distribution is a good example.
# Parameters for the Poisson distribution (our population distribution)
lambda_population = 5  # Average rate of events in the population

# Generate a large population dataset from the Poisson distribution
population_size = 100000
population_data = poisson.rvs(mu=lambda_population, size=population_size)

# Parameters for the CLT simulation
sample_size = 30  # Sample size (should be >= 30 for CLT to apply well)
num_samples = 1000 # Number of samples to draw from the population

# List to store the means of each sample
sample_means = []

# Draw multiple samples and calculate their means
for _ in range(num_samples):
  # Randomly draw a sample of size sample_size from the population
  sample = np.random.choice(population_data, size=sample_size, replace=True)
  # Calculate the mean of the sample
  sample_mean = np.mean(sample)
  # Append the sample mean to the list
  sample_means.append(sample_mean)

# Convert the list of sample means to a NumPy array for easier plotting
sample_means = np.array(sample_means)

# Plot the histogram of the sample means
plt.figure(figsize=(10, 6))
# Use density=True to normalize the histogram, so the area under the bars is 1
plt.hist(sample_means, bins=30, density=True, edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title(f'Distribution of Sample Means from Poisson Distribution (Sample Size={sample_size}, Num Samples={num_samples})')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# We can also verify that the mean of the sample means is close to the population mean
mean_of_sample_means = np.mean(sample_means)
population_mean = lambda_population # For a Poisson distribution, the mean is lambda

print(f"Population Mean (lambda): {population_mean}")
print(f"Mean of Sample Means: {mean_of_sample_means}")

# The standard deviation of the sample means (Standard Error) should be approximately
# the population standard deviation divided by the square root of the sample size.
# For a Poisson distribution, the variance is lambda, so the standard deviation is sqrt(lambda).
population_std_dev = np.sqrt(lambda_population)
theoretical_standard_error = population_std_dev / np.sqrt(sample_size)
standard_deviation_of_sample_means = np.std(sample_means)

print(f"Theoretical Standard Error: {theoretical_standard_error}")
print(f"Standard Deviation of Sample Means: {standard_deviation_of_sample_means}")

# As you can see, the histogram of sample means tends to be bell-shaped (normal),
# and the mean of the sample means is close to the population mean.
# The standard deviation of the sample means is also close to the theoretical standard error.
# This visually demonstrates the Central Limit Theorem for a non-normal distribution.
```

```python
# prompt: Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
# distribution

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculates the Z-score for a single data point.

    Args:
        data_point: The individual data point.
        mean: The mean of the dataset.
        std_dev: The standard deviation of the dataset.

    Returns:
        The calculated Z-score.
    """
    if std_dev == 0:
        return float('inf') if data_point != mean else 0  # Handle the case of zero standard deviation
    return (data_point - mean) / std_dev

# Example data and parameters
data_point_of_interest = 75
dataset_mean = 70
dataset_std_dev = 5

# Calculate the Z-score
z_score = calculate_z_score(data_point_of_interest, dataset_mean, dataset_std_dev)

print(f"The data point is: {data_point_of_interest}")
print(f"The mean of the dataset is: {dataset_mean}")
print(f"The standard deviation of the dataset is: {dataset_std_dev}")
print(f"The calculated Z-score is: {z_score:.2f}")

# To compare it to a standard normal distribution, we can look up the probability
# of getting a Z-score less than or equal to the calculated Z-score using the CDF.
# We can use the scipy.stats.norm.cdf function for this.

# Calculate the cumulative probability (CDF) for the Z-score
cumulative_probability = norm.cdf(z_score)

print(f"The cumulative probability for a Z-score of {z_score:.2f} is: {cumulative_probability:.4f}")

# Interpretation:
# The cumulative probability tells us the proportion of values in a standard normal
# distribution that are less than or equal to our calculated Z-score.
# For example, a Z-score of 1.96 has a cumulative probability of approximately 0.975,
# meaning about 97.5% of values in a standard normal distribution are less than or equal to 1.96.

# We can also visualize the Z-score on the standard normal distribution plot
def plot_z_score_on_normal(z_score):
    """
    Plots the standard normal distribution and highlights the position of a given Z-score.

    Args:
        z_score: The Z-score to highlight.
    """
    # Define the range of x values for the plot
    x = np.linspace(-4, 4, 100)  # Extend the range to see more of the tails

    # Calculate the PDF values for the standard normal distribution
    pdf_values = norm.pdf(x, loc=0, scale=1)

    # Plot the standard normal distribution PDF
    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf_values, color='darkblue', linewidth=2, label='Standard Normal Distribution (N(0, 1))')

    # Highlight the position of the Z-score
    plt.axvline(z_score, color='red', linestyle='--', linewidth=1.5, label=f'Z-score = {z_score:.2f}')

    # Add labels and title
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.title('Z-score on Standard Normal Distribution')

    # Add a grid for better readability
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()

    # Shade the area to the left of the Z-score (representing the cumulative probability)
    x_shaded = np.linspace(-4, z_score, 100)
    pdf_shaded = norm.pdf(x_shaded, loc=0, scale=1)
    plt.fill_between(x_shaded, pdf_shaded, color='red', alpha=0.2, label=f'P(Z <= {z_score:.2f}) = {norm.cdf(z_score):.4f}')

    plt.show()

# Plot the calculated Z-score on the standard normal distribution
plot_z_score_on_normal(z_score)

# Example with a different data point
data_point_of_interest_2 = 60
z_score_2 = calculate_z_score(data_point_of_interest_2, dataset_mean, dataset_std_dev)
print(f"\nFor data point {data_point_of_interest_2}:")
print(f"The calculated Z-score is: {z_score_2:.2f}")
print(f"The cumulative probability is: {norm.cdf(z_score_2):.4f}")
plot_z_score_on_normal(z_score_2)
```

```python
# prompt:  Implement hypothesis testing using Z-statistics for a sample dataset

from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05):
  """
  Performs a one-sample Z-test for hypothesis testing.

  Args:
    sample_mean: The mean of the sample.
    population_mean: The hypothesized mean of the population (from the null hypothesis).
    population_std_dev: The standard deviation of the population (assumed to be known).
    sample_size: The size of the sample.
    alpha: The significance level (default is 0.05).

  Returns:
    A dictionary containing:
      - 'z_score': The calculated Z-statistic.
      - 'p_value': The calculated p-value for a two-tailed test.
      - 'conclusion': The conclusion of the hypothesis test ('Reject Null Hypothesis' or 'Fail to Reject Null Hypothesis').
  """
  if sample_size <= 0:
    return {"error": "Sample size must be positive."}
  if population_std_dev <= 0:
      return {"error": "Population standard deviation must be positive."}

  # Calculate the standard error of the mean
  standard_error = population_std_dev / np.sqrt(sample_size)

  # Calculate the Z-statistic
  z_score = (sample_mean - population_mean) / standard_error

  # Calculate the two-tailed p-value
  # The p-value is the probability of observing a test statistic as extreme as,
  # or more extreme than, the observed value, assuming the null hypothesis is true.
  # For a two-tailed test, we consider both tails of the distribution.
  p_value = 2 * (1 - norm.cdf(abs(z_score)))

  # Compare the p-value to the significance level (alpha) to make a conclusion
  if p_value < alpha:
    conclusion = "Reject Null Hypothesis"
  else:
    conclusion = "Fail to Reject Null Hypothesis"

  return {
      'z_score': z_score,
      'p_value': p_value,
      'conclusion': conclusion
  }

# --- Example Usage ---

# Sample Data Parameters
observed_sample_mean = 68
sample_n = 50

# Population Parameters (based on the null hypothesis)
hypothesized_population_mean = 70
known_population_std_dev = 10  # Assuming the population standard deviation is known

# Significance level
significance_level = 0.05

# Perform the Z-test
test_results = z_test(observed_sample_mean, hypothesized_population_mean, known_population_std_dev, sample_n, significance_level)

# Print the results
print("--- Z-Test Results ---")
print(f"Observed Sample Mean: {observed_sample_mean}")
print(f"Sample Size: {sample_n}")
print(f"Hypothesized Population Mean (Null Hypothesis): {hypothesized_population_mean}")
print(f"Known Population Standard Deviation: {known_population_std_dev}")
print(f"Significance Level (alpha): {significance_level}")
print(f"Calculated Z-score: {test_results['z_score']:.4f}")
print(f"Calculated p-value (two-tailed): {test_results['p_value']:.4f}")
print(f"Conclusion: {test_results['conclusion']}")

# --- Another Example ---

# Sample Data Parameters
observed_sample_mean_2 = 155
sample_n_2 = 100

# Population Parameters (based on the null hypothesis)
hypothesized_population_mean_2 = 150
known_population_std_dev_2 = 20  # Assuming the population standard deviation is known

# Significance level
significance_level_2 = 0.01

# Perform the Z-test
test_results_2 = z_test(observed_sample_mean_2, hypothesized_population_mean_2, known_population_std_dev_2, sample_n_2, significance_level_2)

# Print the results
print("\n--- Another Z-Test Example Results ---")
print(f"Observed Sample Mean: {observed_sample_mean_2}")
print(f"Sample Size: {sample_n_2}")
print(f"Hypothesized Population Mean (Null Hypothesis): {hypothesized_population_mean_2}")
print(f"Known Population Standard Deviation: {known_population_std_dev_2}")
print(f"Significance Level (alpha): {significance_level_2}")
print(f"Calculated Z-score: {test_results_2['z_score']:.4f}")
print(f"Calculated p-value (two-tailed): {test_results_2['p_value']:.4f}")
print(f"Conclusion: {test_results_2['conclusion']}")
```

```python
# prompt: Create a confidence interval for a dataset using Python and interpret the result

# Let's assume we have a sample dataset of numerical values.
# We will generate some sample data for demonstration purposes.
sample_data = [12, 15, 18, 21, 24, 27, 30, 33, 36, 39]

# Convert the sample data to a NumPy array
sample_data = np.array(sample_data)

# Parameters for calculating the confidence interval
confidence_level = 0.95 # 95% confidence level
alpha = 1 - confidence_level # Significance level

# Calculate the sample mean
sample_mean = np.mean(sample_data)

# Calculate the sample standard deviation
# We use ddof=1 for sample standard deviation (unbiased estimator)
sample_std_dev = np.std(sample_data, ddof=1)

# Get the sample size
n = len(sample_data)

# For a sample size >= 30 or if the population standard deviation is known,
# we can use the Z-distribution for the confidence interval.
# Since our sample size is less than 30 and we don't have the population std dev,
# we should technically use the t-distribution. However, the prompt asks for a
# confidence interval for a dataset, implying we can use the Z-distribution
# if we assume the data is from a normally distributed population or the sample size is large enough for CLT.
# For simplicity and to address the prompt, we will proceed with the Z-distribution
# as it's often used as an approximation with larger sample sizes.
# For a more rigorous approach with small samples, the t-distribution is preferred.

# Calculate the Z-critical value for the desired confidence level
# For a two-tailed interval, we need the Z-score that leaves alpha/2 in each tail.
z_critical = norm.ppf(1 - alpha / 2)

# Calculate the standard error of the mean (SEM)
standard_error = sample_std_dev / np.sqrt(n)

# Calculate the margin of error
margin_of_error = z_critical * standard_error

# Calculate the confidence interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print("--- Confidence Interval Calculation ---")
print(f"Sample Data: {sample_data}")
print(f"Sample Size (n): {n}")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std_dev:.4f}")
print(f"Confidence Level: {confidence_level}")
print(f"Significance Level (alpha): {alpha:.2f}")
print(f"Z-critical value ({confidence_level*100}% confidence): {z_critical:.4f}")
print(f"Standard Error of the Mean (SEM): {standard_error:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"Confidence Interval: ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f})")

# --- Interpretation ---
print("\n--- Interpretation ---")
print(f"We are {confidence_level*100}% confident that the true population mean")
print(f"lies within the interval ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f}).")

# What this means:
# If we were to take many random samples of the same size from the same population
# and calculate a confidence interval for each sample in the same way, we would
# expect that approximately {confidence_level*100}% of those intervals would
# contain the true population mean.

# It does NOT mean there is a {confidence_level*100}% probability that the true
# population mean falls within THIS specific calculated interval. The true population
# mean is a fixed value (though unknown), and the interval either contains it or it doesn't.
# The probability is associated with the method of constructing the interval.
```

```python
# prompt: Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

# Choose parameters for the normal distribution
mean = 50  # True mean of the population
std_dev = 10 # True standard deviation of the population

# Generate data from a normal distribution (representing a sample)
sample_size = 100
generated_data = np.random.normal(mean, std_dev, sample_size)

print("--- Generating Data ---")
print(f"Generated a sample of size {sample_size} from a Normal Distribution")
print(f"  (Theoretical Population Mean: {mean}, Theoretical Population Std Dev: {std_dev})")
print("First 10 data points:", generated_data[:10])

# Calculate sample statistics
sample_mean = np.mean(generated_data)
# Use ddof=1 for sample standard deviation (unbiased estimator)
sample_std_dev = np.std(generated_data, ddof=1)
n = len(generated_data)

print("\n--- Sample Statistics ---")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std_dev:.4f}")
print(f"Sample Size (n): {n}")

# Define the confidence level
confidence_level = 0.95 # For a 95% confidence interval
alpha = 1 - confidence_level # Significance level

# Since we have a sufficiently large sample size (n=100 >= 30), we can use the Z-distribution.
# If the sample size were small and the population standard deviation unknown,
# the t-distribution would be more appropriate.

# Find the Z-critical value for the chosen confidence level
# For a two-tailed interval, we need the Z-score that leaves alpha/2 in each tail.
# norm.ppf(q) gives the inverse of the CDF (percentile function),
# returning the z-score for a given cumulative probability q.
# For a 95% CI, alpha = 0.05, alpha/2 = 0.025. We need the z-score for 1 - 0.025 = 0.975
z_critical = norm.ppf(1 - alpha / 2)

# Calculate the Standard Error of the Mean (SEM)
# SEM = Sample Standard Deviation / sqrt(Sample Size)
standard_error = sample_std_dev / np.sqrt(n)

# Calculate the Margin of Error (ME)
# ME = Z-critical value * Standard Error
margin_of_error = z_critical * standard_error

# Calculate the Confidence Interval
# CI = Sample Mean ¬± Margin of Error
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print("\n--- Confidence Interval Calculation ---")
print(f"Confidence Level: {confidence_level}")
print(f"Significance Level (alpha): {alpha}")
print(f"Z-critical value ({confidence_level*100}% confidence): {z_critical:.4f}")
print(f"Standard Error of the Mean (SEM): {standard_error:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"Confidence Interval: ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f})")

# --- Interpretation ---
print("\n--- Interpretation of the Confidence Interval ---")
print(f"Based on our sample data, we are {confidence_level*100}% confident that the true population mean")
print(f"of the normal distribution from which this sample was drawn lies within the interval:")
print(f"  [{confidence_interval_lower:.4f}, {confidence_interval_upper:.4f}]")

# What this interpretation means:
# This interval provides a range of plausible values for the true population mean.
# It does NOT mean there is a {confidence_level*100}% probability that the true population mean
# falls within THIS specific interval. The true population mean is a fixed value
# (though unknown), and this particular interval either contains it or it doesn't.
# The {confidence_level*100}% confidence refers to the method itself: if we were to
# repeat the sampling process and confidence interval calculation many times,
# approximately {confidence_level*100}% of the intervals constructed would
# contain the true population mean.
```

```python
# prompt: Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

# Parameters for the Poisson distribution
lambda_param = 3  # Average rate of events

# Define a range of possible outcomes (number of occurrences) for the Poisson distribution
# Since it's discrete, we consider integer values. The range should be large enough
# to cover most of the distribution's probability. We can use values up to
# a few times lambda.
possible_outcomes = np.arange(0, int(lambda_param * 4) + 1)

# Calculate the Cumulative Distribution Function (CDF) for each possible outcome
# using the scipy.stats.poisson.cdf function
# mu is the lambda parameter
cdf_values = poisson.cdf(possible_outcomes, mu=lambda_param)

# Print the possible outcomes and their corresponding CDF values
print("--- Poisson CDF Calculation ---")
print(f"Lambda (average rate): {lambda_param}")
print("Possible Outcomes (Number of Occurrences):", possible_outcomes)
print("Cumulative Probabilities (CDF):", cdf_values)

# Plot the Cumulative Distribution Function (CDF)
plt.figure(figsize=(10, 6))
# We use plt.step for discrete distributions
plt.step(possible_outcomes, cdf_values, where='post', color='purple', linewidth=2)

# Add labels and title
plt.xlabel('Number of Occurrences (k)')
plt.ylabel('Cumulative Probability P(X ‚â§ k)')
plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution (lambda={lambda_param})')
plt.xticks(possible_outcomes) # Ensure x-axis ticks are at integer values
plt.yticks(np.arange(0, 1.1, 0.1)) # Ensure y-axis ticks show probability increments
plt.grid(axis='both', linestyle='--', alpha=0.7)

# Show the plot
plt.show()

# --- Interpretation ---
print("\n--- Interpretation of the Poisson CDF ---")

# The CDF value for a specific integer 'k' represents the probability that a random
# variable following this Poisson distribution takes a value less than or equal to 'k'.
# P(X ‚â§ k) = CDF(k)

# Example Interpretations:
example_k1 = 2
cdf_at_k1 = poisson.cdf(example_k1, mu=lambda_param)
print(f"The probability of observing {example_k1} or fewer occurrences is: P(X ‚â§ {example_k1}) = {cdf_at_k1:.4f}")

example_k2 = 5
cdf_at_k2 = poisson.cdf(example_k2, mu=lambda_param)
print(f"The probability of observing {example_k2} or fewer occurrences is: P(X ‚â§ {example_k2}) = {cdf_at_k2:.4f}")

# To find the probability of exactly 'k' occurrences, we can use the PMF:
# P(X = k) = Poisson.pmf(k, mu=lambda_param)
# Or, using the CDF:
# P(X = k) = P(X ‚â§ k) - P(X ‚â§ k-1)
# P(X = k) = CDF(k) - CDF(k-1) for k > 0
# P(X = 0) = CDF(0)

example_k3 = 3
pmf_at_k3 = poisson.pmf(example_k3, mu=lambda_param)
cdf_at_k3 = poisson.cdf(example_k3, mu=lambda_param) # Calculate cdf_at_k3
cdf_at_k3_minus_1 = poisson.cdf(example_k3 - 1, mu=lambda_param) if example_k3 > 0 else 0
prob_exactly_k3_from_cdf = cdf_at_k3 - cdf_at_k3_minus_1

print(f"The probability of observing exactly {example_k3} occurrences (using PMF) is: P(X = {example_k3}) = {pmf_at_k3:.4f}")
print(f"The probability of observing exactly {example_k3} occurrences (using CDF) is: P(X = {example_k3}) = {prob_exactly_k3_from_cdf:.4f}")

# To find the probability of more than 'k' occurrences:
# P(X > k) = 1 - P(X ‚â§ k) = 1 - CDF(k)
example_k4 = 4
prob_more_than_k4 = 1 - poisson.cdf(example_k4, mu=lambda_param)
print(f"The probability of observing more than {example_k4} occurrences is: P(X > {example_k4}) = {prob_more_than_k4:.4f}")

# To find the probability between two values (inclusive):
# P(a ‚â§ X ‚â§ b) = P(X ‚â§ b) - P(X ‚â§ a-1) = CDF(b) - CDF(a-1)
example_a = 2
example_b = 5
prob_between_ab = poisson.cdf(example_b, mu=lambda_param) - poisson.cdf(example_a - 1, mu=lambda_param)
print(f"The probability of observing between {example_a} and {example_b} occurrences (inclusive) is: P({example_a} ‚â§ X ‚â§ {example_b}) = {prob_between_ab:.4f}")

# In summary, the CDF allows us to calculate the probability of a Poisson random
# variable falling within a specific range of values. The step-wise nature of the
# plot reflects the discrete nature of the Poisson distribution, where probability
# accumulates at specific integer values.
```

```python
# prompt:  Implement Z-score normalization on a dataset and visualize its transformation

def z_score_normalize(data):
  """
  Performs Z-score normalization on a dataset.

  Args:
    data: A list or NumPy array of numerical data.

  Returns:
    A NumPy array containing the Z-score normalized data.
  """
  if not isinstance(data, np.ndarray):
    data = np.array(data)

  mean = np.mean(data)
  std_dev = np.std(data)

  if std_dev == 0:
      print("Standard deviation is zero. Z-score normalization cannot be applied.")
      return np.zeros_like(data) # Return an array of zeros or the original data

  normalized_data = (data - mean) / std_dev
  return normalized_data

# --- Example Usage ---

# Assume 'generated_data' is available from the preceding code
# If not, you can generate some data here:
# import numpy as np
# generated_data = np.random.normal(50, 10, 100) # Example data

# Perform Z-score normalization
z_scored_data = z_score_normalize(generated_data)

print("Original Data (first 10):", generated_data[:10])
print("Z-score Normalized Data (first 10):", z_scored_data[:10])

# Verify the mean and standard deviation of the normalized data (should be close to 0 and 1)
mean_normalized = np.mean(z_scored_data)
std_dev_normalized = np.std(z_scored_data)

print(f"\nMean of Normalized Data: {mean_normalized:.6f}")
print(f"Standard Deviation of Normalized Data: {std_dev_normalized:.6f}")

# --- Visualization ---

plt.figure(figsize=(12, 6))

# Plot histogram of original data
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.hist(generated_data, bins=20, density=True, edgecolor='black', alpha=0.7, label='Original Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Distribution of Original Data')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Plot histogram of Z-score normalized data
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
plt.hist(z_scored_data, bins=20, density=True, edgecolor='black', alpha=0.7, label='Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.title('Distribution of Z-score Normalized Data')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()

# Interpretation of the visualization:
# The shape of the distribution remains the same after Z-score normalization.
# However, the data is shifted so that its mean is centered around 0, and it is
# scaled so that its standard deviation is 1. This makes it easier to compare
# values from different datasets that originally had different scales and means.
```

```python
# prompt: Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
# distribution.

from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
  """
  Calculates the skewness and kurtosis of a dataset.

  Args:
    data: A list or NumPy array of numerical data.

  Returns:
    A dictionary containing:
      - 'skewness': The calculated skewness of the data.
      - 'kurtosis': The calculated kurtosis of the data (excess kurtosis).
  """
  if not isinstance(data, np.ndarray):
    data = np.array(data)

  # Calculate skewness using scipy.stats.skew
  # default bias=True calculates biased skewness (N in denominator)
  # Setting bias=False calculates unbiased skewness (N-1 in denominator for sample)
  # For large samples, the difference is minimal. We'll use default biased for consistency
  # with the common definition of population skewness, but unbiased is generally preferred for samples.
  data_skewness = skew(data)

  # Calculate kurtosis using scipy.stats.kurtosis
  # default fisher=True calculates excess kurtosis (kurtosis - 3)
  # default bias=True calculates biased kurtosis (N in denominator)
  # Setting bias=False calculates unbiased kurtosis. Again, using biased default here.
  data_kurtosis = kurtosis(data) # This is excess kurtosis

  return {
      'skewness': data_skewness,
      'kurtosis': data_kurtosis
  }

# --- Example Usage ---

# Generate data from a normal distribution
# Using parameters that were used in the preceding code for context
mean = 50
std_dev = 10
sample_size = 10000 # Use a larger sample size for better approximation of population stats
normal_data = np.random.normal(mean, std_dev, sample_size)

# Calculate skewness and kurtosis for the generated normal data
normal_stats = calculate_skewness_kurtosis(normal_data)

print("--- Skewness and Kurtosis for Data from Normal Distribution ---")
print(f"Generated sample of size {sample_size} from N(mean={mean}, std_dev={std_dev})")
print(f"Calculated Skewness: {normal_stats['skewness']:.6f}")
print(f"Calculated Kurtosis (Excess): {normal_stats['kurtosis']:.6f}")

# --- Interpretation for Normal Distribution ---
print("\n--- Interpretation ---")
print("For a perfect normal distribution:")
print("  - Skewness is 0 (symmetric).")
print("  - Kurtosis (Excess) is 0 (mesokurtic).")
print("Our calculated values are close to 0, as expected for data generated from a normal distribution.")

# --- Example with a Skewed Distribution (e.g., Chi-squared) ---
# You can uncomment this to see the difference for a non-normal distribution
# from scipy.stats import chi2
# df = 5 # Degrees of freedom
# skewed_data = chi2.rvs(df, size=sample_size)
# skewed_stats = calculate_skewness_kurtosis(skewed_data)

# print("\n--- Skewness and Kurtosis for Data from Chi-squared Distribution ---")
# print(f"Generated sample of size {sample_size} from Chi-squared (df={df})")
# print(f"Calculated Skewness: {skewed_stats['skewness']:.6f}")
# print(f"Calculated Kurtosis (Excess): {skewed_stats['kurtosis']:.6f}")

# Interpretation for Skewed Distribution:
# Positive skewness means the tail is longer on the right side.
# Positive excess kurtosis means the tails are heavier (more outliers) and the peak is sharper than a normal distribution.


# --- Example with a Different Normal Sample ---
mean_2 = 100
std_dev_2 = 5
sample_size_2 = 5000
normal_data_2 = np.random.normal(mean_2, std_dev_2, sample_size_2)

normal_stats_2 = calculate_skewness_kurtosis(normal_data_2)

print("\n--- Skewness and Kurtosis for Another Normal Sample ---")
print(f"Generated sample of size {sample_size_2} from N(mean={mean_2}, std_dev={std_dev_2})")
print(f"Calculated Skewness: {normal_stats_2['skewness']:.6f}")
print(f"Calculated Kurtosis (Excess): {normal_stats_2['kurtosis']:.6f}")
print("\nThe values are again close to 0, reinforcing the expected shape for normal distribution data.")
```
